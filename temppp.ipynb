```python
import pandas as pd
import json


file = r"C:\Users\khale\Desktop\Geospatial Analysis\Github\spatial-wealth-distribution\Processing Codes\all_admin2_boundaries.json"
with open(file, 'r', encoding='utf-8-sig') as json_file:
    data = json.load(json_file)
```

```python
data = pd.DataFrame(data)
```

```python
import re

new_regions = ['أسيوط', 'الفيوم', 'بورسعيد', 'المنيا', 'الدقهلية', 'القليوبية', 'الاسكندرية']
boundaries_interest = data[data['ADM1'].str.contains('|'.join(map(re.escape, new_regions)), regex=True)]
```

```python
locations_data = []
for index, row in boundaries_interest.iterrows():
    location = {
        "ADM2": row['ADM2'].strip(),
        "ADM1": row['ADM1'].strip(),
        "lat_range": [row['lat_range'][0], row['lat_range'][1]],
        "long_range": [row['long_range'][0], row['long_range'][1]]
    }
    locations_data.append(location)

data = {
    "locations": locations_data
}

with open('phase2_all_adm2.json', 'w', encoding='utf-8') as outfile:
    json.dump(data, outfile, ensure_ascii=False, indent=4)
```
